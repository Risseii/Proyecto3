```python
import random
import matplotlib.pyplot as plot

num_canicas = 3000
lista_canicas = []

def resultados():
  for i in range(num_canicas): #Se va a generar 3000 veces
    x = 6
    for j in range(12):
        a = random.randint(0,1)
        if a == 1:
          x = x + 1 #Si el numero aleatorio es 1 se va sumando en 1 la x
        else:
          x = x - 1 #Por lo contrario, se va restando en 1 la x
    #print(x)
    lista_canicas.append(x) #Se agrega el valor de x en una lista denominada lista_canicas

resultados()


def grafica():
  intervalos = range(min(lista_canicas),max(lista_canicas))
  plot.hist(x=lista_canicas,bins=intervalos,color='pink')
  plot.title("Simulación de la máquina de Galton")
  plot.xlabel('Distribución')
  plot.xticks(intervalos)
  plot.ylabel('Cantidad de canicas')
  plot.show()

grafica()
  


```
